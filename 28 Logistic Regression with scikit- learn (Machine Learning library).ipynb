```python
#convert churn from (yes,no) to (0,1)
le=preprocessing.LabelEncoder()
le.fit(df['Churn'])
list(le.classes_)
dfn=le.transform(df['Churn'])
```

```python
le.inverse_transform([0,1])
df['Y']=dfn
df.info()
df.columns
```

```python
#separate out independent categorical variables for converting into numerical
x_catg=df.loc[:,('gender','Partner','Dependents','PowerService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod')]
x_catg.head()
x_catg.shape
```

```python
for col in x_catg.columns:    
    if len(x_catg[col].unique().tolist())<=5:
        print ("Unique items in" , col , len(df[col].unique().tolist()))
        print('Count of ',col)
        df.groupby(col).nunique()
```

```python
df_numeric = df.select_dtypes(include=[np.number])
df_numeric.info()
```

```python
df_categorical =df.select_dtypes(exclude=[np.number])
df_categorical.info()
df_categorical.head()
```

```python
# Separate out independent numerical variables
x_num=df.loc[:,('SeniorCitizen','tenure','MonthlyCharges','TotalCharges')]
x_num.head()
```

```python
#Convert Categorical vars to dummy coding
x_num1=pd.get_dummies(x_catg)
x_num1.head()
```

```python
for col in x_num1.columns:    
    if len(x_num1[col].unique().tolist())<=5:
        print ("Unique items in" , col , len(x_num1[col].unique().tolist()))
        print('Count of ',col)
        x_num1.groupby(col).nunique()
```

```python
#Get all Xs together
x=pd.concat([x_num1,x_num],axis=1)
x.head()
```

```python
#Create Dependent Variable
y=df['Y']
y.head()
```

```python
x.describe
```

```python
#Create train and test data 75% and 25%
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=1)
train_x.shape
test_x.shape
train_y.shape
test_y.shape
type(train_x)
```

```python
#Build a logistic regression model
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
log.fit(train_x,train_y)
```

```python
log.coef_
```

```python
#find out key predictor of churn
coeff=pd.concat([pd.DataFrame(x.columns),pd.DataFrame(np.transpose(log.coef_))],axis=1)
coeff.columns=('Variable','Coeff')
coeff.sort_values('Variable',ascending=True)
```
